## Bokeh

```python
import numpy as np
import pandas as pd
import os

from bokeh.io import output_notebook, output_file

from bokeh.layouts import layout
from bokeh.layouts import layout, widgetbox, column, row

from bokeh.models import Axis
from bokeh.models import Legend
from bokeh.models import ColumnDataSource
from bokeh.models import HoverTool

from bokeh.palettes import Set3
from bokeh.charts import Bar, show
```

```python
output_notebook()
```

### Import Stata files 

```python
os.chdir(r"H:\Teacher Labor MRKT Shortage\SASS\Raw data")
df = pd.read_excel("SASS New teachers - IPEDS Completion data.xls")
```

```python
df
```

```python
# Reshape data
long = pd.melt(df, id_vars=['year'], value_vars=['total_teachers', 'new_teachers'])
long
```

```python
# Add new labels for legend
long['legend'] = np.where(long['variable']=='total_teachers', 'Total Education Graduates', 'Teachers New to Profession')
```

```python
long.drop
```

#### Bar graph

```python
source = ColumnDataSource(long)

p = Bar(long, label='year', values='value', group='legend')

p.xaxis.axis_label = 'Year'
p.yaxis.axis_label = 'Number of Individuals'

yaxis = p.select(dict(type=Axis, layout="left"))[0]
yaxis.formatter.use_scientific = False 

p.add_tools(HoverTool(tooltips=[("Year", "@year"), 
                               ("Value", "@height")]))

show(p)
```
